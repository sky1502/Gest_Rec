```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
from tqdm import tqdm
import json
import time
import random
import pickle
```

```python
df=pd.read_csv(r'E:\capstone\code\WLASL-master\WLASL-master\start_kit\data.csv',dtype={'video_id': str})
```

```python
df.head()
```

```python
path=r'E:\capstone\code\WLASL-master\WLASL-master\start_kit\videos'
error_video=[]

for i in tqdm(range(df.shape[0])):
    try:
        
        video_id=df.iloc[i]['video_id']
        gloss=df.iloc[i]['gloss']
        split=df.iloc[i]['split']
        url=video_id+".mp4"
        v_path=os.path.join(path,url)
        if not os.path.isfile(v_path):
            continue
        cap=cv2.VideoCapture(v_path)
        video=[]
        
        while(cap.isOpened()):    
            ret, frame = cap.read()
            if ret==True:
                frame = cv2.resize(frame, (48,64), interpolation = cv2.INTER_AREA)
                video.append(np.asarray(frame))
            else:
                break
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
                
       
        video=np.asarray(video)
        if not os.path.exists('E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64/{}'.format(gloss)):
            os.makedirs('E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64/{}'.format(gloss))
    
        if not os.path.exists('E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64/{}/{}'.format(gloss,split)):
            os.makedirs('E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64/{}/{}'.format(gloss,split))
        key='E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64/{}/{}/{}.npy'.format(gloss,split,video_id)
        np.save(key,video)
        
        
    except Exception as e:
        error_video.append(video_id)
        print("error",e)
        continue
```
