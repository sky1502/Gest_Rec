```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
from tqdm import tqdm
import json
```

```python
file_path = r'E:\capstone\code\WLASL-master\WLASL-master\start_kit\WLASL_v0.3.json'
with open(file_path) as ipf:
    content = json.load(ipf)
```

```python
content[0]['instances'][0]
```

```python
col=['gloss','video_id','xmin', 'ymin', 'xmax', 'ymax','fps','frame_end','frame_start','instance_id','signer_id','source','split','url','variation_id']
```

```python
df=[]
for word_det in tqdm(content):
    gloss=word_det['gloss']
    
    for video_det in word_det['instances']:
        temp=[]
        temp.append(gloss)
        temp.append(video_det['video_id'])
        temp.append(int(video_det['bbox'][0]))
        temp.append(int(video_det['bbox'][1]))
        temp.append(int(video_det['bbox'][2]))
        temp.append(int(video_det['bbox'][3]))
        temp.append(float(video_det['fps']))
        temp.append(int(video_det['frame_end']))
        temp.append(int(video_det['frame_start']))
        temp.append(int(video_det['instance_id']))
        temp.append(int(video_det['signer_id']))
        temp.append(video_det['source'])
        temp.append(video_det['split'])
        temp.append(video_det['url'])
        temp.append(int(video_det['variation_id']))
        '''
        if len(video_det['video_id'])  != 5:
            print(len(video_det['video_id']))
        '''
        df.append(temp)
        
        
        
        
    
```

```python
df=pd.DataFrame(df,columns=col)
```
