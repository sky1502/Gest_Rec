```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
from tqdm import tqdm
import json
```

```python
embeddings_index = dict()
f = open(r'C:\Users\HP\Downloads\glove.840B.300d\glove.840B.300d.txt',encoding="utf8")

for line in tqdm(f):
    values = line.split()
    word = values[0]
    try:
        coefs = np.asarray(values[1:], dtype='float32')
    except:
        print(values)
        print("****************")
    embeddings_index[word] = coefs
```
